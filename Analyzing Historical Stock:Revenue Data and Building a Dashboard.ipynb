### Question 2: Use Webscraping to Extract Tesla Revenue Data
### Display the last five rows of the tesla_revenue dataframe using the tail function. Upload a screenshot of the results.


```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
url = "https://www.macrotrends.net/stocks/charts/TSLA/tesla/revenue"
headers = {"User-Agent": "Mozilla/5.0"}
response = requests.get(url, headers=headers)
soup = BeautifulSoup(response.text, "html.parser")

# Get the 2nd table on the page (Quarterly Revenue)
tables = soup.find_all("table", class_="historical_data_table table")
revenue_table = tables[1]

tesla_revenue = pd.read_html(str(revenue_table))[0]
tesla_revenue.columns = ["Date", "Revenue"]
tesla_revenue = tesla_revenue[tesla_revenue["Revenue"] != ""]
tesla_revenue["Revenue"] = tesla_revenue["Revenue"].replace(
    {"\$": "", ",": ""}, regex=True).astype(float)

print("Tesla Revenue (last 5 rows):")
print(tesla_revenue.tail())
```

### Question 5: Plot Tesla Stock Graph
### Use the make_graph function to graph the Tesla Stock Data, also provide a title for the graph.
### Upload a screenshot of your results.

```python
import matplotlib.pyplot as plt

def make_graph(data, title):
    plt.figure(figsize=(12, 6))
    plt.plot(data['Date'], data['Close'])
    plt.title(title)
    plt.xlabel("Date")
    plt.ylabel("Close Price USD ($)")
    plt.grid(True)
    plt.tight_layout()
    plt.show()

# Plot Tesla
make_graph(tesla_data, "Tesla Stock Price History")
```
